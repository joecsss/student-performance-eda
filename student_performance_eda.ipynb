```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Performance.csv")
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

# **The dataset contains no missing values.**


---



```python
sns.histplot(df['overall_score'], kde=True)
plt.title("Distribution of Overall Scores")
```

# **Most students score around the average, with fewer students at very low and very high scores.**


---



```python
sns.scatterplot(
    x='study_hours',
    y='overall_score',
    data=df
)
plt.title("Study Hours vs Overall Score")
```

# **Students who study more hours tend to achieve higher overall scores, though some exceptions exist.**


---



```python
sns.scatterplot(
    x='attendance_percentage',
    y='overall_score',
    data=df
)
plt.title("Attendance vs Overall Score")
```

# **Higher attendance is generally associated with higher academic performance.**
